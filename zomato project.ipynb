```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe =pd.read_csv("Zomato data .csv")
```

```python
print(dataframe)
```

# convert the data type of columm rating

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

```

```python
dataframe.head()
```

```python
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

# checking null values

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of restaurant")
```

# conclusion-majority of the customers falls in dinning category

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="green",marker=0)
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("votes", c="red", size=20)
```

 # dinning restaurant gets maximum votes

```python
plt.figure(figsize=(6,4))
plt.hist(dataframe['rate'], bins=5, edgecolor='black')
plt.title("Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

# majority restaurants got ratings from 3.5 to 4

# average order spending by couples

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

# conclusion- maximum couples prefer restaurants with an approximate cost of 300 rupees

# which mode(online or offline ) got maximum ratings

```python
plt.figure(figsize= (6,6))
sns.boxplot(x='online_order', y='rate', data=dataframe)
```

# conclusion - online order have higher  rating as compared to offline mode

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap ='YlGnBu', fmt='d')
plt.title("heatmap")
plt.xlabel("online_order")
plt.ylabel("listed_in(type)")
plt.show()
```
